### Q1. Write a python program to extract the video URL of the first five videos.



```python
from selenium import webdriver
from selenium.common import exceptions
import sys
import time
from flask import Flask, render_template, request, jsonify
from flask_cors import CORS,cross_origin
import logging
import os
import requests
import re
import pandas as pd

def scrape(url): 
    driver.get(url)
    driver.maximize_window
    time.sleep(5)
    
if __name__ == "__main__":
    driver = webdriver.Chrome()
    url = 'https://www.youtube.com/@PW-Foundation/videos'
    scrape(url)
```

```python
url_data = driver.find_elements("xpath",'//*[@id="video-title-link"]')
url = []
for i in url_data:
    url.append(i.get_attribute('href'))
print(url[:5])
```

### Q2. Write a python program to extract the URL of the video thumbnails of the first five videos.



```python
input_text = 'PW-Foundation'
url = f'https://www.youtube.com/@{input_text}/videos'

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"}
response = requests.get(url, headers=headers)
print(response.status_code)
response_text = response.text

vid_thumbnails = re.findall(r"https://i.ytimg.com/vi/[A-Za-z0-9_-]{11}/[A-Za-z0-9_]{9}.jpg", response_text)
thumbnail = []

for i in vid_thumbnails:
    if i not in thumbnail:
        thumbnail.append(i)
        
print(thumbnail[:5])
```

### Q3. Write a python program to extract the title of the first five videos.



```python
title_data = driver.find_elements("xpath",'//*[@id="video-title-link"]')
title = []
for i in title_data:
    title.append(i.get_attribute('title'))
print(title[:5])
```

### Q4. Write a python program to extract the number of views of the first five videos.



```python
views_data = driver.find_elements("xpath",'//*[@id="metadata-line"]/span[1]')
views = []
for i in views_data:
    views.append(i.text)
print(views[:5])
```

### Q5. Write a python program to extract the time of posting of video for the first five videos.

```python
upload_data = driver.find_elements("xpath",'//*[@id="metadata-line"]/span[2]')
upload = []
for i in upload_data:
    upload.append(i.text)
print(upload[:5])
```

```python
keys = ['Titles','Thumbnails','Links','Viewcounts','Ages']
details_dict = {'Titles':title[:30],'Thumbnails':thumbnail[:30],'Links':url[:30],'Viewcouts':views[:30],'Ages':upload[:30]}
print(details_dict)
```

```python
df = pd.DataFrame(details_dict)
df.to_csv('YTscrapData.csv', index=False)
```
